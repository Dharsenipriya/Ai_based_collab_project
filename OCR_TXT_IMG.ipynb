```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
IMAGE_PATH = '/content/sample_data/Screenshot 2025-08-14 133022.png'
#IMAGE_PATH = 'surf.jpeg'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, font, 0.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread(IMAGE_PATH)
spacer = 100
for detection in result:
    # Extract bounding box coordinates and convert to integers
    top_left = tuple(map(int, detection[0][0]))
    bottom_right = tuple(map(int, detection[0][2]))
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15

plt.imshow(img)
plt.show()
```

# Task
Generate code to extract text from an image and convert the extracted text to a word document.

## Process extracted text

### Subtask:
Iterate through the `result` variable to extract the text strings.


**Reasoning**:
Initialize an empty list and loop through the result to extract the text strings.



```python
extracted_texts = []
for detection in result:
    text = detection[1]
    extracted_texts.append(text)
```

## Combine text

### Subtask:
Join the extracted text strings into a single, coherent block of text.


**Reasoning**:
Join the extracted text strings into a single string with spaces.



```python
combined_text = ' '.join(extracted_texts)
```

## Write to file

### Subtask:
Write the combined text to a file with a `.doc` or `.docx` extension.


**Reasoning**:
Write the combined text to a file with a `.doc` extension.



```python
with open('extracted_text.doc', 'w') as f:
    f.write(combined_text)
```
